```python
import sweetviz as sv
import pandas as pd
```

```python
file_path = "../Data/Raw Data/titanic - titanic.csv"
```

```python
# Load Titanic dataset
train_df = pd.read_csv("train.csv")

# Generate report
report = sv.analyze(train_df)

# Save report
report.show_html("Titanic_EDA_Report.html")
```

```python
# Load both train and test datasets
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")

# Compare both datasets
compare_report = sv.compare([train_df, "Train"], [test_df, "Test"])

# Save comparison report
compare_report.show_html("Titanic_Train_vs_Test_Comparison.html")
```

```python
# Analyze with target
report_with_target = sv.analyze(train_df, target_feat="Survived")

# Save report
report_with_target.show_html("Titanic_Survival_Feature_Impact.html")
```

```python
# Load Loan dataset
loan_df = pd.read_csv("loan_data.csv")  # Update filename accordingly

# Analyze with target
loan_report = sv.analyze(loan_df, target_feat="Loan_Status")

# Save report
loan_report.show_html("Loan_Status_Feature_Impact.html")
```
